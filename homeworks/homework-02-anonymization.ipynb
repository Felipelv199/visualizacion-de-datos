# Anonimización de bases de datos

```python
import pandas as pd

df_titanic = pd.read_csv("../data/titanic/train.csv")
list(df_titanic.columns)
```

```python
df_titanic
```

```python
import hashlib

def encript_str(s: str):
    hash_object = hashlib.sha224(s.encode('utf-8'))
    return hash_object.hexdigest()

m, n = df_titanic.shape

decrypted_data = {
    "passengerId": [],
    "name": []
}

for i in range(m):
    name = df_titanic.at[i,"Name"]
    passengerId = df_titanic.at[i,"PassengerId"]
    df_titanic.at[i,"Name"] = encript_str(name)
    decrypted_data["passengerId"].append(passengerId)
    decrypted_data["name"].append(name)

df_titanic
```

```python
ages_ranges = []

for i in range(10):
    i_age = i*10
    ages_ranges.append((i_age,i_age+9))

ange_range_list = []

for i in range(m):
    age = df_titanic.at[i,"Age"]

    for age_range in ages_ranges:
        min_range, max_range = age_range
        if not pd.notna(age):
            ange_range_list.append(f"<=Age<=")
            break
        elif min_range <= age <= max_range:
            ange_range_list.append(f"{min_range}<=Age<={max_range}")
            break

age_range_series = pd.Series(ange_range_list, name="Age Range")
df_anonymized = pd.concat([df_titanic, age_range_series],axis=1,ignore_index=False)
df_anonymized.drop("Age",1,inplace=True)
df_anonymized
```
